### Prediction functions

```python
IMG_SIZE = 224
```

```python
# Function to process and predict image
def image_prediction(image, MODEL, CLASS_NAMES, NUM_CLASSES):
    # Process input image
    image_resized = cv2.resize(image, (IMG_SIZE, IMG_SIZE))
    image = np.expand_dims(image_resized, axis=0)
    
    # Threshold
    threshold = (1/NUM_CLASSES) * 1.2 # 120% from uniform split probability as threshold
    
    # Run image through model to obtain prediction
    pred = MODEL.predict(image)
        # pred is a array of length class_count. Each entry is the probability for a particular class
        # Index based on class_names
        
    if pred[0][np.argmax(pred)] < threshold:
        pred_class = "Unknown"
            
    else:    
        # Obtain prediction class
        pred_class = CLASS_NAMES[np.argmax(pred)]
        
    return pred_class
```

```python
def layered_prediction (image):
    general_pred = image_prediction(image, model_general, CLASS_GENERAL, NUM_GENERAL)
    
    if general_pred == 'Garden and Urban Birds':
        final_pred = image_prediction(image, model_garden, CLASS_GARDEN, NUM_GARDEN)
    elif general_pred == 'Kingfishers':
        final_pred = image_prediction(image, model_king, CLASS_KING, NUM_KING)
    elif general_pred == 'Raptors':
        final_pred = image_prediction(image, model_raptor, CLASS_RAPTOR, NUM_RAPTOR)
    elif general_pred == 'Waterbirds':
        final_pred = image_prediction(image, model_water, CLASS_WATER, NUM_WATER)
    elif general_pred == 'Woodpeckers and Barbets':
        final_pred = image_prediction(image, model_wood, CLASS_WOOD, NUM_WOOD)
    
    return final_pred
```
